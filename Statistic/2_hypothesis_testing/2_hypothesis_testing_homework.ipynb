+ Данное наблюдение точно является выбросом, т.к. очень сильно отклоняется от среднего значения
+ Прежде чем судить о том, насколько необычным является это наблюдение, необходимо знать, чему равняется стандартное отклонение

Ответ:
+ Прежде чем судить о том, насколько необычным является это наблюдение, необходимо знать, чему равняется стандартное отклонение
