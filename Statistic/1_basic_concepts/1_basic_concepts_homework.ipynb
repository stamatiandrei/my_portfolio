# Задание 2
___
Соотнесите способы формирования выборки с предложенными примерами.

A – Для того чтобы протестировать курс по введению в статистику и выяснить, насколько хорошо с ним справятся студенты курса по аналитике, случайным образом было выбрано 35 студентов

B – Для исследования взаимосвязи успешности прохождения курса по статистике и бекграунда, студентов разделили на 2 группы программистов и менеджеров. Затем из каждой группы случайным образом извлекли по 30 человек

C – Для проверки знаний студентов с разных онлайн-курсов по анализу данных, случайным образом было выбрано несколько соответствующих курсов, из студентов которых случайным образом было отобрано по 30 человек с разных потоков.

Ответ:
+ С -> Групповая выборка (cluster sample)
+ А -> Простая случайная выборка (random sample)
+ В -> Стратифицированная выборка (stratified sample)
