# Задание 1
___

Два стрелка одновременно стреляют по мишени. Вероятность попадания по мишени у первого стрелка равна 0,5, у второго — 0,8. Какова вероятность того, что в мишени будет только одна пробоина?

### Разбор решения

В данной задаче вам необходимо рассчитать вероятность тех случаев, в которых один стрелок попадет, а другой — промажет. Это можно сделать следующим образом:

1. Рассчитать вероятность того, что оба стрелка попадут в мишень — это будет перемножение вероятностей: 0.5×0.8=0.4 
2. Рассчитать вероятность того, что оба стрелка промажут по мишени — это будет перемножение вероятностей промаха: (1−0.5)×(1−0.8)=0.1
3. Так как у нас суммарная вероятность всех событий равна 1, то мы можем вычесть вероятность этих двух ситуаций и получить вероятность того, что хотя бы один из стрелков попадет в мишень: 1−0.4−0.1=0.5

Также можно пойти другим путем и определить вероятности событий:

1. Первый стрелок попал, а второй стрелок промахнулся: 0.5×(1−0.8)=0.1
2. Ситуация наоборот, первый стрелок промахнулся, а второй попал: (1−0.5)×0.8=0.4
3. Сложить эти вероятности, так как наши ситуация независимы 0.1+0.4=0.5

# Задание 2
___

Трое аналитиков на экзамене независимо друг от друга решают одну и ту же задачу. Вероятности ее решения этими аналитиками равны 0,8, 0,7 и 0,6 соответственно. Найдите вероятность того, что хотя бы один аналитик решит задачу.

### Разбор решения

Решение задачи каждым конкретным аналитиком - событие независимое, поэтому мы можем рассчитать вероятности того, решит каждый аналитик задачку или нет:

$q_1 = 1−p_1 = 1−0.8 = 0.2$ \
$q_2 = 1−p_2 = 1−0.7 = 0.3$ \
$q_3 = 1−p_3 = 1−0.6 = 0.4$ 

Тогда по теореме умножения вероятностей независимых событий: $p_0 = q_1×q_2×q_3 = 0.2×0.3×0.4 = 0.024$ - вероятность того, что будет три промаха. Тогда $1−p_0 = 1−0.024 = 0.976$ – вероятность того, что хотя бы один аналитик сможет решить задачку.

# Задание 3
___
В первой урне находятся 10 белых и 4 черных шаров, а во второй 5 белых и 9 черных шаров. Из каждой урны вынули по шару. Какова вероятность того, что оба шара окажутся черными? В ответ запишите значение, округленное до тысячных.

### Разбор решения

В этой задаче все события независимы друг от друга, поэтому нужно просто правильно определить вероятности:

1. Вероятность вытащить черный шар из первой урны - 4/14
2. Вероятность вытащить черный шар из второй урны - 9/14

Так как события независимые, то нам достаточно просто перемножить полученные значения - 4/14×9/14=0.1836735 и, округляя до тысячных, мы получаем 0.184

# Задание 4
___
Вероятность хотя бы одного попадания в цель при четырех выстрелах равна 0,9984. Найдите вероятность попадания в цель при одном выстреле.
### Разбор решения
Пусть вероятность попадания в цель при одном выстреле будет равна  p. Каждый выстрел — это независимое событие. Тогда вероятность хотя одного попадания в цель при четырех выстрелах равна $1−(1−p)^4$, так как вероятность, что не будет ни одного попадания, равна $(1−p)^4$, и по условию это выражение равно 0,9984. Тогда нам нужно просто решить уравнение выше: \
$(1−p)^4 = 0,0016$ \
$(1−p) = 0,2$ \
$p=0,8$

# Задание 5
___
Из колоды в 36 карт наудачу извлекается одна карта. События: A = {вынутая карта – туз}, В = {вынутая карта чёрной масти}, F = {вынутая карта – фигура, т.е. является валетом, дамой, королём или тузом}. Выберите все пары зависимых событий:

1) B и F \
2) А и В \
3) A и F 
### Разбор решения
+ События B и F будут независимыми, так как вы можете вытянуть любую карту черной масти, а вот окажется ли она "фигурой" — ещё не факт
+ События A и B также будут независимыми, так как вы можете вытянуть туз и красной, и черной масти.
+ Последняя пара событий как раз будет являться зависимой — ведь если вы вытяните туз, то вероятность вытянуть туз в событии F изменится, ведь тузов в колоде станет меньше

# Задание 6
___
В конверте находится 10 лотерейных билетов, среди которых 3 выигрышных. Из конверта последовательно извлекаются билеты. Найти вероятность того, что три первых лотерейных билета будут выигрышными. Ответ округлите до десятитысячных.
### Разбор решения
Вероятность того, что первый вытащенный конверт будет выигрышный — 3/10, вероятность того, что и второй конверт будет выигрышный — 2/9, если один выигрышный уже достали, третий — 1/8. Почему именно так? Потому что общее количество билетов уменьшается с каждым новым взятым билетом.

Так как события независимы, то их вероятности мы перемножаем. Значит, вероятность вытащить три выигрышных билета = 3/10∗2/9∗1/8≈0.0083

# Задание 7
___
В трех углах равностороннего треугольника находится по роботу. Каждый из роботов начинает двигаться в случайно выбранный угол по стороне треугольника. Какова вероятность того, что ни один из роботов не столкнется с другим роботом?
### Разбор решения
Роботы не столкнутся только в том случае, если не пойдут навстречу друг другу. Значит, если роботы расположены в вершинах [A, B, C], то схема переходов должна быть следующая

$A -> B, B -> C, C -> A$

Или же в обратном порядке:

$A -> C, C -> B, B -> A$

Всего количество способов передвижения для роботов = 2^3 = 8. Нам подходят только 2 => ответ = 2/8 = 0.25

# Задание 8
___
Брошены две игральные кости. Есть следующие события:

А = {выпадение шестерки на первой кости} \
В = {сумма выпавших очков равна 7} \
С = {выпадение тройки на второй кости}

Какие пары событий являются независимыми? 
### Разбор решения
Довольно очевидно, что A и C друг от друга не зависят - ведь бросок одной кости никак не меняет результат на другой. Но как быть с A и B, а также B и C? Проверим это через формулу независимости событий:

$P(A∩B)=P(A)P(B)$

Символ ∩ обозначает здесь "И".

P(A)=1/6 (так как всего граней шесть, а нас интересует всего одно значение)

P(B)=6/36=1/6 (всего комбинаций двух кубиков 36, а условию соответствуют варианты 6 и 1, 5 и 2, 4 и 3, 3 и 4, 2 и 5, 1 и 6)

P(A∩B)=1/36 (если выпала шестерка, то событие B произойдёт только для одной комбинации из 36 - 6 и 1)

Итого: 1/36=1/6∗1/6, что соответствует независимости событий. То же самое справедливо для события C.

Значит, все эти события независимы.

Бросаются 2 игральных кубика. Определите вероятность того, что произведение числа очков на обоих кубиках не превосходит 10.

В ответ запишите число округленное до тысячных
### Разбор решения
Всего комбинаций значений на двух кубиках - 36. Нас устраивают из них следующие варианты (произведение которых не больше 10):

(1, 1) (1, 2) ... (1, 6)

(2, 1) (2, 2) ... (2, 5)

(3, 1) (3, 2) (3, 3)

(4, 1) (4, 2)

(5, 1) (5, 2)

(6, 1)

Всего таких вариантов 19 => ответ 19/36
