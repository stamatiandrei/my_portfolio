# Задание 1
___
Из 500 компьютеров  180 принадлежат к 1 партии, 170 – ко второй партии, остальные к третьей. В первой партии 3% брака, во второй — 2%, в третьей – 6%. Случайно выбирается один компьютер. Определить вероятность того, что выбранный компьютер – бракованный. 

### Разбор решения

Давайте найдем вероятность каждого события по классическому определении вероятности. Всего 500 компьютеров, из них первой партии принадлежат 180, то есть $P(H1) = \frac{500}{180} = 0.36$, 2-ой партии принадлежат 170, то есть $P(H2) = \frac{500}{170} = 0.34$, остальные компьютеры принадлежат к третей партии, то есть $P(H3)= {500}{150} = 0.3$.

Введем событие А = (бракованный компьютер). По условию даны априорные вероятности $P(A∣H1)=0.03$, $P(A∣H2)=0.02$, $P(A∣H3)=0.06$

Вероятность события А можем найти по формуле полной вероятности: $P(A)=P(A∣H1)P(H1)+P(A∣H2)P(H2)+P(A∣H3)P(H3)=0.03×0.36+0.02×0.34+0.06×0.3=0.0356$

Вероятность сделать работу за один день для Java-программиста — 0,85, для программиста на языке C++ 0,9, а для Python программиста 0,8. Найти вероятность того, что программист, выбранный наугад, сделает работу за один день. 

### Разбор решения

Давате найдем вероятность того, на каком языке пишет каждый программист, где наименьшую долю выразим через x. Тогда программистов на языке Python - x, программистов на языке C++ - 4x, а програмистов на языке Java - 12x:

$P(H1) = \frac{x}{x+4x+12x} = \frac{1}{17}$ - Python

$P(H2) = \frac{4x}{x+4x+12x} = \frac{4}{17}$ - C++

$P(H3) = \frac{12x}{x+4x+12x} = \frac{12}{17}$ - Java

Введем событие А = (успешно решенная задача). По условию даны априорные вероятности $P(A∣H1)=0.8$, $P(A∣H2)=0.9$,  $P(A∣H3)=0.85$

Вероятность события А можем найти по формуле полной вероятности: $P(A)=P(A∣H1)P(H1)+P(A∣H2)P(H2)+P(A∣H3)P(H3)= \frac{1}{17} × 0.8 + \frac{4}{17} × 0.9 + \frac{12}{17} × 0.85 = 0.86$

# Задание 3
___
Из 40 снайперов 18 попадает в цель с вероятностью 0,9, 8 — с вероятностью 0,4 и 14 – с вероятностью 0,7. Случайно выбранный снайпер произвел выстрел, поразив цель. Мы хотим вычислить, снайпер из какой группы наиболее вероятно совершил этот выстрел. 

### Разбор решения

Сделаем следующее допущение:

$H1$ — снайпер принадлежал первой группе

$H2$ — снайпер принадлежал второй группе

$H3$ — снайпер принадлежал третьей группе

Тогда мы можем по классическому определению вероятности рассчитать:

$P(H1) = \frac{18}{40} = 0.45$ 

$P(H2)= \frac{14}{40} = 0.35$

$P(H3)= \frac{8}{40} = 0.2$

Найдем вероятность события А (снайпер попал в мишень) и введем условные вероятности:

$P(A∣H1)=0.9$

$P(A∣H2)=0.7$

$P(A∣H3)=0.4$

И тогда по формуле полной вероятности:

$P(A)=P(A∣H1)×P(H1)+P(A∣H2)×P(H2)+P(A∣H3)×P(H3)=0.45×0.9+0.35×0.7+0.2×0.4=0.73$

Теперь найдем апостериорные вероятности того, что стрелок принадлежит к конкретной группе, если он попал в цель:

$P(H1∣A) = \frac{P(A∣H1)×P(H1)}{P(A)} = \frac{0.45×0.9}{0.73} = 0.55479$

$P(H2∣A) = \frac{P(A∣H2)×P(H2)}{P(A)} = \frac{0.35×0.7}{0.73} = 0.3356$

$P(H3∣A) = \frac{P(A∣H3)×P(H3)}{P(A)} = \frac{0.2×0.4}{0.73} = 0.10959$

# Задание 4
___
Есть 4 кубика. На трех из них окрашена белым половина граней, а на четвертом кубике всего одна грань из шести белая. Наудачу выбранный кубик подбрасывается семь раз. Найти вероятность того, что был выбран четвертый кубик, если при семи подбрасываниях белая грань выпала ровно один раз.  Ответ округлите до тысячных. Используйте точку в качестве десятичного разделителя.

### Разбор решения

Пусть $H1$ - это событие, когда выбирают 1-ый, 2-ой или 3-ий кубик, а H2 — когда выбирают 4-ый кубик.

Тогда по классическому определению вероятности: $P(H1)= \frac{3}{4} ,P(H2)= \frac{1}{4}$

Для события $H1$ вероятность появления белой грани $p=0.5$.

Посчитаем вероятность того, что при n=7 (семи бросках) белая грань выпадает ровно k=1. Используем формулу Бернулли:

$P_7^1 = C _7^1*0.5^1*0.5^{7−1} =0.0547=P(A∣H1)$

Теперь рассматриваем событие H2

Вероятность появления белой грани $p=\frac{1}{6}$ , тогда вероятность по формуле Бернулли при n=7, а k=1 будет:

$P_7^1=C_7^1*\frac{1}{6}^1*\frac{5}{6}^{7-1}=0.3907=P(A∣H2)$

Тогда по формуле полной вероятности получаем:

$P(A)=P(H1)×P(A∣H1)+P(H2)×P(A∣H2)=0.75×0.0547+0.25×0.3907=0.1387$

И нам остался финальный штрих - уточнить вероятность гипотезы H2

$P(H2∣A)= \frac{P(A∣H2)×P(H2)}{P(A)} = \frac{0.25×0.3907}{0.1387}=0.7042$

# Задание 6
___
Известно, что около 60% всех людей в городе болеют заболеванием А. Для определения того, заболел человек или нет, проводится тест, точность которого составляет 99%. Вы получили положительный тест, какова вероятность того, что вы реально больны?

### Разбор решения
Пусть,
$A$ − событие, которое соответствует болезни человека ,

$H_1$ − событие, которое соответствует положительному результату теста

$H_2$ − событие, которое соответствует отрицательному результату теста

Тогда по формуле Байеса,

$P(A∣H_1)=\frac{P(H_1)P(H_1∣A)}{P(A)}$ \
$P(H_1∣A)=0.99$, так как мы знаем, что если человек болеет, то тест это покажет с 99 % точностью

$P(A)=0.6$ -- вероятность того, что человек болен

По формуле полной вероятности,

$P(H_1)=P(H_1∣A)∗P(A)+P(H_1∣!A)∗P(!A)=0.99∗0.6+0.01∗0.4=0.598$

Подставляем всё в формулу:
$P(A∣H_1)=\frac{0.99 ∗0.6}{0.598}≈0.993$

# Задание 7
___
Три фабрики производят лампочки. Фабрика А производит 20% всех лампочек, 50% лампочек производится на фабрике В и оставшиеся 30% на фабрике С. 2% лампочек, произведенных на фабрике А, 1% лампочек, произведенных на фабрике В, и 3% лампочек, произведенных на фабрике С, являются бракованными.На рынке наугад выбирается лампочка и оказывается бракованной. Какова вероятность того, что эта лампочка была произведена на заводе B?

### Разбор решения

Пусть, $A,B,C$−  события, соответствующие тому, что лампочка с завода $A$, $B$ и $C$ соответственно, а $H$− событие, соответствующее тому, что лампочка бракованная, тогда

$P(B∣H)=\frac{P(H∣B)P(B)}{P(H)}$

$P(H∣B)=0.01$

$P(B)=0.5$

$P(H)=P(H∣A)∗P(A)+P(H∣B)∗P(B)+P(H∣C)∗P(C)=0.02∗0.2+0.01∗0.5+0.03∗0.3=0.018$

$P(B∣H)=\frac{0.01∗0.5}{0.018}≈0.278$

# Задание 8
___
Система РЛС (радиолокационная система) может обнаружить присутствие самолета в 98% случаев. Однако если в зоне действия радара нет самолета, система все равно сообщает (ложно), что самолет присутствует, в 5% случаев. В любой момент времени вероятность того, что самолет находится в зоне действия радара, составляет 7%.Какова вероятность того, что самолет находится в зоне действия радара, учитывая, что на радаре нет самолета?

### Разбор решения

$A$− событие, соответствующее тому, что самолет в зоне

$B$− событие, соответствующее тому, что самолет не обнаружен

Тогда вероятность того, что самолет не будет обнаружен, даже если он в зоне:

$P(B∣A)=1−0.98=0.02$

А вероятность того, что самолет не обнаружат, если его реально нет в зоне:

$P(B∣A‘)=1−0.05=0.95$

Воспользуемся теоремой Байеса:

$P(A∣B)=\frac{P(B∣A)∗P(A)}{P(B∣A)∗P(A)+P(B∣A‘)∗P(A‘)}=\frac{0.02∗0.07}{0.02∗0.07+0.95∗0.93}=0.0016$

$P(B∣A)=0.98$

# Задание 10
___
Передатчик аварийного локатора самолета (ELT) - это устройство, предназначенное для передачи сигнала в случае аварии. Случайно выбранный ELT был протестирован и оказался бракованным. Найдите вероятность того, что он был произведен компанией A.

### Разбор решения

Пусть, $A,B,C$ −  события, соответствующие тому, что ELT от компании $A$, $B$ и $C$ соответственно, а $H$ − событие соответствующее тому, что ELT бракованный, тогда

$P(A∣H)=\frac{P(H∣A)P(A)}{P(H)}$

$P(H∣A)=0.04$

$P(A)=0.8$

$P(H)=P(H∣A)∗P(A)+P(H∣B)∗P(B)+P(H∣C)∗P(C)=0.04 ∗0.8+0.06 ∗0.15+0.09 ∗0.05= 0.0455$

$P(A∣H)=\frac{0.04∗0.8}{0.0455}≈0.703$
