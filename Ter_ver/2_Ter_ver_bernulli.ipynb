# Задание 1
___
Устройство, состоящее из пяти независимо работающих элементов, включается за время Т. Вероятность отказа каждого из них за это время равна 0,4. Найти вероятность того, что откажут три элемента.

### Разбор решения:
В данном случае мы можем вспомнить формулу Бернулли:
$P^n_k = C_n^kp^kq^{n−k}$, где $C_n^k = \frac{n!}{k!(n−k)!}$ \
Тогда мы получаем, что p (вероятность отказа элемента) = 0.4, а число испытаний (т.е. число элементов) n=5.

Вероятность того, что откажут ровно три элемента из пяти: $P_5(3)=C_5^3 × 0.4^3 × 0.6^2 ≈ 0.2304$ 

Где $C_5^3$ рассчитывается как:

```python
from math import factorial
```

```python
def num_of_successes(n, k):
    return factorial(n)/(factorial(k) * factorial(n - k))
```

```python
num_of_successes(5, 3)
```

### Разбор решения
В данном случае мы применяем формулу Бернулли:
$P^n_k = C_n^kp^kq^{n−k}$, где $C_n^k = \frac{n!}{k!(n−k)!}$

Тогда мы получаем, что p (вероятность попадания в цель) =0.3, а число испытаний (т.е. число выстрелов) n=8. А число попаданий в цель в данном случае будет k=2.

Теперь рассмотрим ситуацию, что мы хотим попасть хотя бы два раза, т.е. нас не устраивают случаи, когда мы совсем не попадаем (k=0) или попадаем 1 раз (k=1).

Подставляем значения в формулу Бернулли:

$1 − C_8^1 × 0.3^1 × (1−0.3)^{8−1} − C_8^0 × 0.3^0 × (1−0.3)^8 = 0.745$

В магазине 7 покупателей. Каждый может совершить покупку с вероятностью 0,4. Найти вероятность того, что не более двух человек совершат покупку. Ответ округлите до сотых. В качестве десятичного разделителя используйте точку.

### Разбор решения
В данном случае мы применяем формулу Бернулли:
$P^n_k = C_n^kp^kq^{n−k}$, где $C_n^k = \frac{n!}{k!(n−k)!}$
 

Тогда мы получаем, что p (вероятность совершения покупки) =0.4, а число испытаний (т.е. число покупателей) n=7. 

Нас здесь интересуют три варианта: когда никто не совершил покупку (k=0), когда её совершил только один покупатель (k=1) и когда её совершило два покупателя (k=2).

Подставляем значения в формулу Бернулли:

$C_7^0 × 0.4^0 × (1−0.4)^7 + C_7^1 × 0.4^1 × (1−0.4)^{7−1} + C_7^2 × 0.4^2 × (1−0.4)^{7−2} ≈ 0.42$

# Задание 5

В телеателье имеется 7 телевизоров. Для каждого телевизора вероятность того, что в данный момент он включен, равна 0,5. Найти вероятность того, что в данный момент включены не менее трех телевизоров.

### Разбор решения
В данном случае мы применяем формулу Бернулли:
$P^n_k = C_n^kp^kq^{n−k}$, где $C_n^k = \frac{n!}{k!(n−k)!}$

Тогда мы получаем, что p (вероятность того, что телевизор включен) =0.5, а число испытаний (т.е. число телевизоров) n=7. 

Здесь нам не подходят только три варианта: когда не включён ни один телевизор (k=0), когда включён только один (k=1) и когда включены только два (k=2). Соответственно, через обратную вероятность мы сможем вычислить всё остальное.

Подставляем значения в формулу Бернулли:

$1 − C_7^0 × 0.5^0 × (1−0.5)^7 − C_7^1 × 0.5 × (1−0.5)^{7−1} − C_7^2 × 0.5^2 × (1−0.5)^{7−2} = 0.773$

# Задание 6
___
Пусть вероятность того, что покупателю необходим компьютер с операционной системой Linux, равна 0.25. Найдите вероятность того, что из шести покупателей по крайней мере двум необходим компьютер с операционной системой Linux.

### Разбор решения
Воспользуемся понятием отрицательной вероятности, т. е. $P(A) = 1 - P(\bar{A})$

В нашем случае, A - такое событие, что из шести покупателей по крайней мере двум необходим компьютер с операционной системой Linux

Соответственно, не А - такое событие, что одному покупателю необходим компьютер или ни одному покупателю не необходим компьютер. В таком случае:

$1 - (P(одному\_покупателю\_необходим\_компьютер\_c\_Linux) + P(никому\_не\_необходим\_компьютер\_с\_Linux)) = 1− C_6^1*0.25^1*0.75^5 + C_6^0*0.25^0*0.75^6 ≈ 0.47$
